```python
df = pd.read_csv('student_data.csv')
df.head()
```

```python
sns.lineplot(data = df, x='study_hours', y='exam_score')
#data mein df dalenge x axis mein study_hours y axis mein exam_score

plt.tight_layout()
plt.show()
```

```python
def get_grade(score):
    if score >= 90:
        return 'A'
    elif score >= 80:
        return 'B'
    elif score >= 70:
        return 'C'
    elif score >= 60:
        return 'D'
    else:
        return 'F'


df['grade'] = df['exam_score'].apply(get_grade)
df.head()
#
```

```python
sns.barplot(data=df, x='grade', y='exam_score', estimator = 'mean')

plt.title('Exam Score by Grade')
# plt.xlabel('Grade')
# plt.ylabel('Exam Score')
plt.tight_layout()
plt.show()
#
```
