```python
class student:
  def __init__(self, name, age):
    self.name = name
    self.age = age



  def display(self):

    print("Name: ", self.name)
    print("Age: ", self.age)


a = student('John', 20)

a.display()



print(a.name)      #output: John
print(a.age)       #output: 20
```
