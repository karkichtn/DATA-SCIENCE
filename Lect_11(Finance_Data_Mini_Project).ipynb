##good loans, bad loans


```python
good_loans = df[df['loan_status'] == 'Fully Paid']
bad_loans = df[df['loan_status'] != 'Fully Paid']

print(f'Good loans: {len(good_loans)}')
print(f'Bad loans: {len(bad_loans)}')
```

```python
#as_index = false loan_status' ke baad mein lagana hain


loan_status_data = df.groupby('loan_status').agg({
    'id' : 'count',
    'dti' : 'mean',
    'int_rate' : 'mean',
    'loan_amount' : 'sum',
    'total_payment' : 'sum'
})

loan_status_data.rename(columns = {'id' : 'Number of Application'}, inplace = True)

# loan_status_data['Number of Application'] = loan_status_data['Number of Application'].astype(int)



display(loan_status_data)
```

##good loan analysis (teacher)




```python
#good loan application


no_of_applications = loan_status_data.loc['Current', 'Number of Application'] + loan_status_data.loc['Fully Paid', 'Number of Application']
print(no_of_applications)



#good loan funded amount


funded_amount = loan_status_data.loc['Current', 'loan_amount'] + loan_status_data.loc['Fully Paid', 'loan_amount']
print(funded_amount)



#good loan total received amount



received_amount = loan_status_data.loc['Current', 'total_payment'] + loan_status_data.loc['Fully Paid', 'total_payment']
print(received_amount)



#profit

profit = received_amount - funded_amount
print(profit)


```

##good loan analysis(self)

```python
#good loan application
#good loan application percentage
#good loan funded amount
#good loan total received amount






#good loan application percentage
good_loans = df[df['loan_status'] == 'Fully Paid' ]
Number_of_Good_Loans = len(good_loans)
Total_Number_of_Loans = len(df)

Good_Loan_Application_Percentage = (Number_of_Good_Loans / Total_Number_of_Loans) * 100


print(f'Good Loan Application Percentage: {Good_Loan_Application_Percentage:.2f}%')




#good loan funded amount
good_loan_funded_amount = good_loans['loan_amount'].sum()
print(f'Good Loan Funded Amount : {good_loan_funded_amount}')



#good loan total received amount
good_loan_total_received_amount = good_loans['total_payment'].sum()
print(f'Good Loan Toal Received Amount : {good_loan_total_received_amount}')
```

##bad loan analysis


```python
#bad loan application percentage
#bad loan applications
#bad loan funded amount
#bad loan toal received amount





#bad loan application percentage

bad_loans = df[df['loan_status'] != 'Fully Paid']
Number_of_Bad_Loans = len(bad_loans)
Total_Number_of_Loans = len(df)

Bad_Loan_Application_Percentage = (Number_of_Bad_Loans / Total_Number_of_Loans) * 100

print(f'Bad Loan Application Percentage: {Bad_Loan_Application_Percentage:.2f}%')
```
