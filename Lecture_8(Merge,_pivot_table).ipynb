```python
import pandas as pd

df1 = pd.DataFrame({
    'id':[1,2,3,4],
    'name':['Raju','Rani','Rahul','Rhea'],
    'age':[25,30,35,40]
})

df2 = pd.DataFrame({
    'id':[3,4,5,6],
    'name':['Rahul','Rhea','Ravi','Reena'],
    'age':[35,40,45,50]

})


df3 = pd.merge(df1, df2, on='id', how='inner')   #inner joint


df3.rename(columns={'name_x':'table_name1','age_x':'age'},inplace=True)      #used for renameing the column name

display(df3)
print(" ")



df4 = pd.merge(df1, df2, on='id', how='left')   #left joint
display(df4)
print(" ")


df5 = pd.merge(df1, df2, on='id', how='right')   #right joint
display(df5)
print(" ")


df6 = pd.merge(df1, df2, on='id', how='outer')   #outer joint
display(df6)
print(" ")


df7 = pd.merge(df1, df2, how='cross')   #cross joint
display(df7)
print(" ")


df8 = pd.concat([df1,df2])               #concat dono table ko upar niche join kar deta hain
display(df8)
print(" ")


df9 = pd.concat([df1,df2],axis=1)         #dono table ko side by side join kar dega with every column mtlb id 2 baar aayega table mein bhi
display(df9)
print(" ")



```

```python
df1 = pd.DataFrame({
    'id':[1,2,3,4,1,1,3,4],
    'name':['Raju','Rani','Rahul','Rhea','Raju','Raju','Rahul','Rhea'],
    'age':[25,30,35,40,25,25,35,40]
})

df2 = pd.DataFrame({
    'id':[3,4,5,6],
    'name':['Rahul','Rhea','Ravi','Reena'],
    'age':[35,40,45,50]

})


df3 = pd.merge(df1, df2, on='id', how='inner')

#drop_duplicates duplicate value ko remove kar deta hain or agar uske parameter mein subset mein id likhenge toh woh sirf id jo unique hogi usko print karega agar koi parameter nahi hoga toh
#jo jo equal hain puri row mein unko element karega or 1 hi baar likhega


#drop_duplicates() duplicate rows ko remove karne ke liye use hota hai. Hum decide kar sakte hain kis
#column ke basis par duplicate check karna hai aur kaunsi duplicate row rakhni hai.

df3.drop_duplicates(subset = ['id'], keep='first', inplace = True)
display(df3)
```

```python
#pivot table concept

#Pivot table ek table ki values ko apne se place karta hain mtlb agar column mein hume not placed placed likhna hain or uske ander study hours ka mean nikalna hain toh uske liye use hota hain



#Pivot Table: Data ko summarize karne ke liye use hoti hai.
#Ek column ki values ko rows ya columns mein arrange karke kisi
#numeric value ka summary (mean, sum, count, etc.) nikalti hai.
```

```python
import pandas as pd

df = pd.read_csv('student_data.csv')
```

```python
df.pivot_table(columns = 'placement_status', index = 'sleep_hours', values = 'exam_score', aggfunc = 'mean')
```
