```python
lst = [1, 2, 3, 'hello', 'c', [2, 4]]

print(lst)

#append
lst.append([2,3,4])
print(lst)


#extend()
lst.extend([2, 3, 4])
print(lst)


#insert()
lst.insert(2, "value")
print(lst)


#pop
lst.pop()          #last vala element hatega bas or agar andar koi number dalega toh uska index vala element hat jayega
print(lst)

#remove
lst.remove(2)      #jo pahale vala 2 hoga woh hatega bas
print(lst)

#clear
lst.clear()
print(lst)


#index()
lst = [1, 2, 3, 'hello', 'c', [2, 4]]
print(lst.index('hello'))

#count()
print(lst.count(2))


#sort()
lst = ['hello', 'uber', 'isagi']          #ya toh pura int le list mein ya string ko, mix mat kar
lst.sort()
print(lst)

#reverse()
lst.reverse()
print(lst)
```

```python
#sum, min ,max, sorted


#sum
lst=[2, 3 ,4453, 23, 422, -4312, 3412]
print(sum(lst))

#min
print(min(lst))

#max
print(max(lst))

#sorted
print(sorted(lst))

print(lst)
```

```python
#tuple

tp=(1,2,33,4, 2, 33, 5, 3, 53, 23, 2)
a = tuple(sorted(tp))
print(a)


#tuple function

#Category           Functions
#count values       count()
#Find               index()
#size               len()
#total              sum()
#highest/lowest     max()/min()
#sort              sorted()(return list)




print(tp.count(2))
print(tp.index(33))
print(len(tp))
print(sum(tp))
print(max(tp))
print(min(tp))
print(sorted(tp))
```

```python
lst1=[]
lst2=[]

size = int(input())

for i in range(size):
    lst1.append(int(input()))
    if(i==size-1):
      print(lst1)


for i in range(size):
    lst2.append(int(input()))
    if(i==size-1):
      print(lst1)
      print(lst2)


lst3 =[]
for i in range(len(lst1)):
    lst3.append(lst1[i] + lst2[i])


print(lst3)
```
